```python
number = int(input("enter a number:"))

if number > 10:
    print(f"{number} is greater than 10")
elif number == 10:
    print(f"{number} is equal to 10")
else:
    print(f"{number} is less than 10")
```

#### Task 1

```python
age = int(input("Enter Age:"))

if 0 < age <=5:
    print(f"You have stated you are {age} years old. you get Free Entry, Welcome")
elif 5 < age <= 12:
    print(f"You have stated you are {age} years old. Please pay $5 Entry")
elif 12 < age <= 17:
    print(f"You have stated you are {age} years old. Please pay $8 Entry")
elif 17 < age <= 59:
    print(f"You have stated you are {age} years old. Please pay $12 Entry")
elif 59 < age <= 116:
    print(f"You have stated you are {age} years old. you get Free Entry, Welcome")
elif age > 116:
    print("You are waaayyy to old to be going to the cinema")
elif age < 0:
    print("You need to be born first to come to the cinema")
else:
    Print("Invalid Response")
```

#### Task 2

```python
grade = int(input("enter percentage:"))

if grade > 100:
    print("Grade is invalid")
elif 90 <= grade < 100:
    print(f"Grade A Achieved, {grade}%")
elif 80 <= grade < 90:
    print(f"Grade B Achieved, {grade}%")
elif 70 <= grade < 80:
    print(f"Grade C Achieved, {grade}%")
elif 60 <= grade < 70:
    print(f"Grade D Achieved, {grade}%")
elif 0 <= grade < 60:
    print("Assessment Failed")
elif grade < 0:
    print("Grade is invalid")
```
